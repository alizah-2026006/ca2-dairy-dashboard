```python
milk_production= pd.read_csv("milk_production.csv")
milk_production.head()
```

```python
# Filter the dataframe to contain only Ireland data
milk_production_ie = milk_production[(milk_production["geo"] == "Ireland") & (milk_production["dairyprod"]== "Raw milk, total available on farms") & 
    (milk_production["milkitem"]== "Products obtained (1 000 t)")]
milk_production_ie.head()
```

```python
milk_production_ie.shape
```

```python
milk_production_ie.info()
```

```python
milk_production_ie.describe(include='object')
```

Dropping the irrelevant features based on dataset knowledge and previous outputs.

```python
milk_production_ie = milk_production_ie.drop(['DATAFLOW','LAST UPDATE','freq','dairyprod','milkitem','geo',
     'OBS_FLAG','CONF_STATUS'], axis=1).reset_index(drop=True)

milk_production_ie.head()
```

```python
milk_production_ie = milk_production_ie.rename(columns={'TIME_PERIOD': 'year', 'OBS_VALUE':'milk_production (1000 t)'})
milk_production_ie.head()
```

```python
processed_dairy = pd.read_csv("processed_dairy_production.csv")
processed_dairy.head()
```

```python
processed_dairy.info()
```

```python
processed_dairy.describe(include="object")
```

```python
processed_dairy["Product"].unique()
```

```python
# Converting long format to wide format using "pivot"
processed_dairy = processed_dairy.pivot(index="Year",values="VALUE",columns="Product").reset_index()
processed_dairy.columns.name = None  # Resetting column index name
processed_dairy=processed_dairy.rename(columns={"Year":"year","Butter":"butter_production (t)","Cheese":"cheese_production (t)",
                                                "Skimmed Milk Powder":"skimmed_milk_powder(t)"})
processed_dairy.head()
```
