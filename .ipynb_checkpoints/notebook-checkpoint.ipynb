```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings('ignore') # supress the warnings
```

```python
milk_production = pd.read_csv("milk_production.csv")
```

```python
milk_production.head()
```

```python
milk_production.info()
```

```python
milk_production['geo'].unique()
```

```python
milk_production['dairyprod'].unique()
```

```python
# Filter the dataframe to contain only Ireland data
milk_production_ie = milk_production[(milk_production["geo"] == "Ireland")]
milk_production_ie.head()
```

```python
milk_production_ie.shape
```

```python
milk_production_ie.info()
```

```python
milk_production_ie.describe(include='object')
```

Dropping the irrelevant features based on dataset knowledge and previous outputs.

```python
milk_production_ie = milk_production_ie.drop(['DATAFLOW','LAST UPDATE','freq','dairyprod','milkitem','geo',
     'OBS_FLAG','CONF_STATUS'], axis=1).reset_index(drop=True)

milk_production_ie.head()
```

```python
milk_production_ie
```

```python
milk_production_ie = milk_production_ie.rename(columns={'TIME_PERIOD': 'year', 'OBS_VALUE':'milk_production (thousand tonnes)'})
milk_production_ie.head()
```

```python
milk_production_ie.isnull().sum()
```
